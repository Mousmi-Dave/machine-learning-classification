```python
#Import required Library & Classes
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#Read_csv file
df=pd.read_csv('Social_Network_Ads.csv')
df.head()
```

```python
df.info() #Checking null values
```

```python
df['Purchased'].value_counts() #Will show how many purchases has been done & not 
```

```python
gender=pd.get_dummies(df['Gender'],drop_first=True)
gender.columns=['Gender']
df.drop('Gender', axis=1, inplace=True)
df=pd.concat([df,gender],axis=1)
```

```python
df.head()
```

```python
X=df[['Age','EstimatedSalary','Gender']] #Independent Data
y=df['Purchased'] #Dependent Data
```

```python
Sf=StandardScaler() #standardize features
scaled=Sf.fit_transform(X)
```

```python
dff=pd.DataFrame(scaled,columns=X.columns)
dff.head()
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(scaled,y,test_size=0.2,random_state=42)
```

```python
KNN=KNeighborsClassifier(n_neighbors=1)
KNN.fit(Xtrain,ytrain)
```

```python
pred=KNN.predict(Xtest)
```

```python
confusion_matrix(pred,ytest,labels=[0,1])
```

```python
print(classification_report(ytest,pred))
```
